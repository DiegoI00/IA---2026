# 1. Regresión Lineal

## Descripción general

La **Regresión Lineal** es un algoritmo de aprendizaje supervisado utilizado para predecir valores numéricos continuos. Su objetivo es encontrar una relación lineal entre una o varias variables de entrada y una variable de salida.

Por ejemplo, puede utilizarse para predecir el precio de una casa, las ventas de una empresa, la temperatura o cualquier valor cuantitativo.


## Fundamento matemático

La Regresión Lineal busca ajustar una función de la forma:

$$
\hat{y} = w_1x_1 + w_2x_2 + \dots + w_nx_n + b
$$

También puede escribirse como:

$$
\hat{y} = w^T x + b
$$

Donde:

- $x$ representa las variables de entrada.
- $w$ representa los pesos o coeficientes del modelo.
- $b$ representa el sesgo o intercepto.
- $\hat{y}$ es el valor predicho por el modelo.
- $y$ es el valor real.

El objetivo del algoritmo es encontrar los valores de $w$ y $b$ que produzcan el menor error posible entre las predicciones y los valores reales.

---

## Función de costo

La función de costo más utilizada en Regresión Lineal es el Error Cuadrático Medio, conocido como MSE que es el promedio de :

$$
J(w,b) = \frac{1}{m}\sum_{i=1}^{m}(\hat{y}^{(i)} - y^{(i)})^2
$$

Donde:

- $m$ es el número total de ejemplos de entrenamiento.
- $\hat{y}^{(i)}$ es la predicción del ejemplo $i$.
- $y^{(i)}$ es el valor real del ejemplo $i$.

El algoritmo busca minimizar esta función de costo.

---

## Descripción formal

Dado un conjunto de entrenamiento:

$$
D = \{(x^{(1)}, y^{(1)}), (x^{(2)}, y^{(2)}), \dots, (x^{(m)}, y^{(m)})\}
$$

la Regresión Lineal aprende una función:

$$
f(x) = w^T x + b
$$

de manera que el error entre los valores reales y los valores predichos sea el menor posible.

---

## Pseudocódigo

```text
Algoritmo Regresión Lineal

Entrada:
    X: matriz de características
    y: valores reales
    alpha: tasa de aprendizaje
    epochs: número de iteraciones

Inicializar:
    w = vector de pesos en cero o aleatorio
    b = 0

Para cada época desde 1 hasta epochs:

    Calcular predicción:
        y_pred = X * w + b

    Calcular error:
        error = y_pred - y

    Actualizar pesos:
        w = w - alpha * promedio(error * X)

    Actualizar sesgo:
        b = b - alpha * promedio(error)

Fin Para

Salida:
    w, b

# Naive Bayes

## 1. Descripción general

**Naive Bayes** es un algoritmo de aprendizaje supervisado utilizado para problemas de **clasificación**. Se basa en el **Teorema de Bayes** y calcula la probabilidad de que un dato pertenezca a una clase determinada.

Se utiliza con frecuencia en problemas como:

- Clasificación de correos spam.
- Clasificación de texto.
- Análisis de sentimientos.
- Diagnóstico médico.
- Clasificación de documentos.

Es un modelo probabilístico porque toma decisiones con base en probabilidades.

---

## 2. Fundamento matemático

Naive Bayes se basa en el **Teorema de Bayes**, el cual se expresa de la siguiente manera:

$$
P(C|X) = \frac{P(X|C)P(C)}{P(X)}
$$

Donde:

- $C$ representa una clase.
- $X$ representa el conjunto de características observadas.
- $P(C|X)$ es la probabilidad de que el dato pertenezca a la clase $C$, dado que se observaron las características $X$.
- $P(X|C)$ es la probabilidad de observar las características $X$, dado que el dato pertenece a la clase $C$.
- $P(C)$ es la probabilidad previa de la clase.
- $P(X)$ es la probabilidad de observar las características.

El objetivo del algoritmo es calcular la probabilidad de cada clase y seleccionar aquella que tenga el mayor valor.

---

## 3. Supuesto de independencia

El término **Naive** significa “ingenuo”. El algoritmo recibe este nombre porque asume que las características son independientes entre sí, dado el valor de la clase.

Es decir, si un dato tiene varias características:

$$
X = (x_1, x_2, x_3, \dots, x_n)
$$

Naive Bayes asume que:

$$
P(X|C) = P(x_1|C)P(x_2|C)P(x_3|C)\dots P(x_n|C)
$$

Por lo tanto, la probabilidad puede expresarse como:

$$
P(C|X) \propto P(C)\prod_{j=1}^{n}P(x_j|C)
$$

Como $P(X)$ es igual para todas las clases, no es necesario calcularlo para comparar las probabilidades. Por eso, el modelo elige la clase que maximiza la siguiente expresión:

$$
\hat{C} = \arg\max_C P(C)\prod_{j=1}^{n}P(x_j|C)
$$

---

## 4. Descripción formal

Dado un conjunto de entrenamiento:

$$
D = \{(x^{(1)}, y^{(1)}), (x^{(2)}, y^{(2)}), \dots, (x^{(m)}, y^{(m)})\}
$$

donde cada $y^{(i)}$ representa la clase correspondiente a cada ejemplo, Naive Bayes aprende:

1. La probabilidad previa de cada clase:

$$
P(C)
$$

2. La probabilidad de cada característica dado cada clase:

$$
P(x_j|C)
$$

3. Para un nuevo dato, calcula la probabilidad de pertenecer a cada clase.

4. Finalmente, selecciona la clase con mayor probabilidad.

---

## 5. Tipos comunes de Naive Bayes

### 5.1 Gaussian Naive Bayes

Se utiliza cuando las variables son numéricas continuas. Asume que los valores de cada característica siguen una distribución normal o gaussiana.

La probabilidad se calcula con la siguiente fórmula:

$$
P(x_j|C) =
\frac{1}{\sqrt{2\pi\sigma_C^2}}
e^{-\frac{(x_j-\mu_C)^2}{2\sigma_C^2}}
$$

Donde:

- $\mu_C$ es la media de la característica para la clase $C$.
- $\sigma_C^2$ es la varianza de la característica para la clase $C$.
- $x_j$ es el valor de la característica que se desea evaluar.

---

### 5.2 Multinomial Naive Bayes

Se utiliza principalmente en problemas de texto, donde las características representan conteos o frecuencias.

Por ejemplo:

- Número de veces que aparece una palabra en un documento.
- Frecuencia de palabras en un correo electrónico.
- Conteo de términos en una reseña.

Este tipo de Naive Bayes es común en clasificación de documentos y detección de spam.

---

### 5.3 Bernoulli Naive Bayes

Se utiliza cuando las características son binarias, es decir, cuando solo pueden tomar valores como 0 o 1.

Por ejemplo:

- Una palabra aparece o no aparece en un documento.
- Una característica está presente o ausente.
- Una respuesta es sí o no.

---

## 6. Pseudocódigo

```text
Algoritmo Naive Bayes

Entrada:
    X: matriz de características
    y: clases reales

Entrenamiento:

    Obtener lista de clases únicas

    Para cada clase C:

        Calcular probabilidad previa:
            P(C) = número de ejemplos de clase C / total de ejemplos

        Para cada característica x_j:

            Calcular P(x_j | C)

            Si los datos son numéricos:
                calcular media y varianza de x_j para la clase C

            Si los datos son categóricos o texto:
                calcular frecuencia de x_j dentro de la clase C

Clasificación de un nuevo dato x:

    Para cada clase C:

        Calcular:
            score(C) = P(C)

        Para cada característica x_j del nuevo dato:

            Actualizar:
                score(C) = score(C) * P(x_j | C)

    Seleccionar la clase con mayor score:
        clase_predicha = clase con mayor score(C)

Salida:
    clase_predicha
```

---

## 7. Resumen

**Naive Bayes** es un algoritmo probabilístico de clasificación basado en el Teorema de Bayes. Calcula la probabilidad de que un dato pertenezca a cada clase y selecciona la clase más probable. Su principal supuesto es que las características son independientes entre sí dado el valor de la clase. Aunque este supuesto no siempre se cumple en la realidad, el algoritmo suele funcionar bien en muchos problemas prácticos, especialmente en clasificación de texto.

# Análisis de complejidad de Naive Bayes

## Variables utilizadas

Para analizar la complejidad del algoritmo Naive Bayes, se consideran las siguientes variables:

- $m$: número de ejemplos de entrenamiento.
- $n$: número de características o atributos.
- $C$: número de clases posibles.
- $V$: número de valores posibles por característica, en caso de datos categóricos o texto.

---

## 1. Complejidad durante el entrenamiento

Durante el entrenamiento, Naive Bayes calcula principalmente dos tipos de probabilidades:

1. La probabilidad previa de cada clase:

$$
P(C)
$$

2. La probabilidad condicional de cada característica dado cada clase:

$$
P(x_j|C)
$$

Para calcular estas probabilidades, el algoritmo recorre los datos de entrenamiento y cuenta la frecuencia de aparición de las clases y de las características asociadas a cada clase.

Si hay $m$ ejemplos y cada ejemplo tiene $n$ características, el algoritmo necesita revisar cada característica de cada ejemplo.

Por lo tanto, la complejidad temporal del entrenamiento es:

$$
O(m \cdot n)
$$

Esto significa que el tiempo de entrenamiento crece de manera proporcional al número de ejemplos y al número de características.

---

## 2. Complejidad durante la clasificación

Para clasificar un nuevo dato, Naive Bayes calcula la probabilidad de que ese dato pertenezca a cada clase.

Para cada clase, evalúa las $n$ características del nuevo ejemplo.

Si existen $C$ clases y cada ejemplo tiene $n$ características, la complejidad temporal para clasificar un solo dato es:

$$
O(C \cdot n)
$$

Si se desean clasificar $t$ nuevos ejemplos, la complejidad sería:

$$
O(t \cdot C \cdot n)
$$

---

## 3. Complejidad espacial

Naive Bayes necesita almacenar:

1. Las probabilidades previas de cada clase:

$$
P(C)
$$

2. Las probabilidades condicionales de cada característica dado cada clase:

$$
P(x_j|C)
$$

En el caso general, si hay $C$ clases, $n$ características y cada característica puede tomar $V$ valores diferentes, la complejidad espacial es:

$$
O(C \cdot n \cdot V)
$$

En el caso de **Gaussian Naive Bayes**, para cada clase y cada característica se almacenan principalmente la media y la varianza. Por lo tanto, la complejidad espacial es:

$$
O(C \cdot n)
$$

---

## 5. Interpretación

Naive Bayes es considerado un algoritmo eficiente porque su entrenamiento suele ser rápido. Solo necesita contar frecuencias o calcular estadísticas simples, como medias y varianzas.

Además, la clasificación también es eficiente, ya que únicamente calcula probabilidades para cada clase y compara cuál es la mayor.

Por esta razón, Naive Bayes es útil cuando se tienen muchos datos, muchas características o problemas de clasificación de texto, donde puede haber una gran cantidad de palabras como atributos.

---

## Conclusión

La complejidad temporal de entrenamiento de Naive Bayes es:

$$
O(m \cdot n)
$$

y la complejidad para clasificar un nuevo dato es:

$$
O(C \cdot n)
$$

Esto lo convierte en un algoritmo rápido y práctico para tareas de clasificación, especialmente cuando se trabaja con grandes volúmenes de datos.

### Problema 1: Métricas de matriz de confusión

Datos:

- TP = 50
- FP = 10
- FN = 5
- TN = 35
- Total = 100

Procedimiento:

Accuracy:

$$
Accuracy = \frac{TP + TN}{TP + TN + FP + FN}
$$

$$
Accuracy = \frac{50 + 35}{100} = 0.85
$$

Resultado:

**Accuracy = 85%**

Precisión:

$$
Precision = \frac{TP}{TP + FP}
$$

$$
Precision = \frac{50}{50 + 10} = 0.833
$$

Resultado:

**Precision = 83.3%**

Recall:

$$
Recall = \frac{TP}{TP + FN}
$$

$$
Recall = \frac{50}{50 + 5} = 0.909
$$

Resultado:

**Recall = 90.9%**

F1-Score:

$$
F1 = 2 \cdot \frac{Precision \cdot Recall}{Precision + Recall}
$$

$$
F1 = 2 \cdot \frac{0.833 \cdot 0.909}{0.833 + 0.909} = 0.869
$$

Resultado:

**F1-Score = 86.9%**

Conclusión:

En un contexto médico, la métrica más importante suele ser el **Recall**, porque interesa detectar la mayor cantidad posible de casos positivos reales. Un falso negativo puede ser grave porque una persona enferma podría no recibir tratamiento.

Datos:

- TP = 8

- FN = 2
- TN = 400

Procedimiento:

$$

Resultado:

Recall:

$$

Resultado:

Precisión:

$$

Resultado:

**Precision ≈ 16.7%**

Conclusión:

El modelo parece bueno por su Accuracy, pero en realidad genera muchos falsos positivos. En datasets desbalanceados, el Accuracy puede ser engañoso. Para medicina, el Recall suele ser más importante porque ayuda a reducir falsos negativos.

### Problema 3: MAE y MSE

Errores:

- Error 1 = 10

Procedimiento MAE:

$$

Resultado:

**MAE = 240**

Resultado:

**MSE = 163,400**

El MSE penaliza mucho más los errores grandes porque eleva los errores al cuadrado. Por eso es más sensible a valores atípicos.

### Problema 4: Interpretación de regresión lineal

Modelo:

$$

- Intercepto = 5: valor esperado de `y` cuando `x1` y `x2` valen 0.
- Coeficiente de `x1` = 2: si `x1` aumenta una unidad, `y` aumenta 2 unidades.
- Coeficiente de `x2` = -3: si `x2` aumenta una unidad, `y` disminuye 3 unidades.

Conclusión:

`x1` tiene relación positiva con la variable objetivo, mientras que `x2` tiene relación negativa. Si `x2` representa tiempo de espera, entonces mayor tiempo de espera reduce la satisfacción.

### Problema 5: Regresión logística y función sigmoide

Modelo:

$$
P(y=1) = \frac{1}{1 + e^{-(2x - 4)}}
$$

Para `x = 2`:

$$
2(2) - 4 = 0
$$

$$
P(y=1) = \frac{1}{1 + e^0} = \frac{1}{2} = 0.5
$$

Resultado:

**Cuando x = 2, la probabilidad es 0.5**

Regla de decisión:

- Si `x > 2`, el modelo predice clase 1.
- Si `x < 2`, el modelo predice clase 0.

Conclusión:

El coeficiente 2 controla la pendiente de la curva. El sesgo -4 desplaza la curva. El punto de decisión está en `x = 2`.

### Problema 7: Kernel en SVM

Problema:

Un SVM lineal falla cuando los datos no son linealmente separables.

Solución:

Usar un kernel para proyectar los datos a un espacio de mayor dimensión.

Fórmula general:

$$
K(x_i, x_j) = \phi(x_i) \cdot \phi(x_j)
$$

Kernel RBF:

$$
K(x,y) = e^{-\frac{||x-y||^2}{2\sigma^2}}
$$

Beneficios:

- Permite separar datos no lineales.
- Mantiene eficiencia computacional.
- Puede usar kernels polinomiales, RBF o sigmoides.

Conclusión:

El truco del kernel permite encontrar fronteras curvas o complejas sin transformar explícitamente los datos.

### Problema 9: Escalamiento en KNN

KNN usa distancias euclidianas. Si una variable tiene escala mucho mayor, domina la distancia.

El ingreso domina la clasificación y la edad casi no influye.

$$
x' = \frac{x - min(x)}{max(x) - min(x)}
$$

Solución 2: Estandarización Z-score

$$
z = \frac{x - \mu}{\sigma}
$$

Conclusión:

Escalar las variables permite que todas contribuyan de forma equilibrada al cálculo de distancias.

$$

Conclusión:

La entropía es alta porque las clases están relativamente mezcladas.

### Problema 4: Profundidad y sobreajuste en árboles

Un árbol con profundidad 20 es muy complejo.

Problema principal:

**Overfitting**

Consecuencias:

- Memoriza los datos de entrenamiento.
- Tiene buen desempeño en entrenamiento.
- Falla con datos nuevos.
- Pierde capacidad de generalización.

Random Forest reduce este problema mediante:

- Bagging.
- Selección aleatoria de características.
- Votación entre árboles.

Si eliminamos la aleatoriedad:

- Todos los árboles serían iguales.
- No habría diversidad.
- No se reduciría la varianza.
- El modelo heredaría el sobreajuste del árbol individual.

Conclusión:

La diversidad de los árboles es lo que permite que Random Forest sea más robusto.

### Problema 5: Gradient Boosting y residuos

Datos:

- Valores reales: 12, 8, 15
- Predicción inicial: 10, 10, 10

Cálculo de residuos:

$$
Residuo = y_{real} - y_{pred}
$$

Resultados:

- Para 12:

$$
12 - 10 = 2
$$

- Para 8:

$$
8 - 10 = -2
$$

- Para 15:

$$
15 - 10 = 5
$$

Vector de residuos:

$$
[2, -2, 5]
$$

Interpretación:

El siguiente árbol aprende los errores del modelo anterior.

Actualización:

$$
NuevaPrediccion = PrediccionActual + \eta(ResiduoPredicho)
$$

Ejemplo con $\eta = 0.1$:

$$
10 + (0.1)(2) = 10.2
$$

Conclusión:

Gradient Boosting corrige errores de forma secuencial. Cada nuevo árbol intenta mejorar lo que el anterior no explicó.

### Problema 6: Random Forest vs Gradient Boosting

Random Forest:

- Entrena árboles en paralelo.
- Cada árbol es independiente.
- Promedia o vota resultados.
- Reduce varianza.
- Es robusto frente al ruido.
- Suele ser más estable.

Gradient Boosting:

- Entrena árboles de forma secuencial.
- Cada árbol corrige errores del anterior.
- Reduce sesgo.
- Puede lograr mayor precisión.
- Es más sensible a ruido y outliers.
- Puede sobreajustar si no se regula.

Conclusión:

- Si se busca un modelo rápido, estable y robusto: **Random Forest**.
- Si se busca máxima precisión y se pueden ajustar hiperparámetros: **Gradient Boosting**.

### Problema 10: Z-score e Isolation Forest

Datos:

$$
[10, 12, 11, 13, 100]
$$

Media:

$$
\bar{x} = \frac{10 + 12 + 11 + 13 + 100}{5} = 29.2
$$

Varianza:

$$
\sigma^2 = 1254.16
$$

Desviación estándar:

$$
\sigma = \sqrt{1254.16} = 35.41
$$

Z-score para 100:

$$
z = \frac{100 - 29.2}{35.41} = 1.99
$$

Z-score para 12:

$$
z = \frac{12 - 29.2}{35.41} = -0.48
$$

Resultado:

- 100 está mucho más alejado del grupo.
- Su Z-score no supera 3 porque el dataset es pequeño y el valor 100 arrastra la media y la desviación estándar.

Isolation Forest:

- Aísla puntos mediante cortes aleatorios.
- Los outliers se aíslan con pocas particiones.
- Los puntos normales requieren más particiones.

Conclusión:

El valor 100 puede considerarse un outlier porque está claramente separado del resto de los datos. En Isolation Forest tendría una profundidad corta, lo que indica anomalía.
